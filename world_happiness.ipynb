```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_columns = ['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
df = data[data_columns].copy()
```

```python
df.head()
```

```python
happiness_dict = {'Country name':'country_name','Regional indicator':'regional_indicator','Ladder score':'happiness_score','Logged GDP per capita':'GDP_per_capita','Social support':'social_support','Healthy life expectancy':'healthy_life_expectancy','Freedom to make life choices':'freedom_to_make_life_choices','Generosity':'generosity','Perceptions of corruption':'perception_of_corruption'}
```

```python
df.rename(columns=happiness_dict, inplace=True)
```

```python
df.columns
```

```python
df.columns.nunique()
```

```python
df.head()
```

```python
df.regional_indicator.unique()
```

```python
df.regional_indicator.nunique()
```

```python
plt.figure(figsize=(12,7))
plt.title('Happiness Score and GDP per Capita', fontsize=40)
sns.scatterplot(x=df.happiness_score, y=df.GDP_per_capita, hue=df.regional_indicator, s=150)
plt.xlabel('Happiness Score', fontsize=20)
plt.ylabel('GDP per Capita', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(loc='lower right')
plt.savefig('happinessgdp.png', bbox_inches='tight')
plt.show()
```

```python
regional_gdp = df.groupby('regional_indicator')['GDP_per_capita'].sum()
regional_gdp
```

```python
plt.figure(figsize=(12,7))
plt.title('Regional GDP Per Capita', fontsize=40)
regional_gdp.plot.pie(autopct= '%1.1f%%')
plt.ylabel('')
plt.show()
```

```python
countries_number = df.groupby('regional_indicator')['country_name'].nunique()
countries_number
```

```python
cor = df.corr(method='pearson')
plt.figure(figsize=(12,7))
sns.heatmap(cor, cmap='coolwarm', annot=True, square=True)
plt.show()
```

```python
top_10 = df.head(10)
```

```python
plt.figure(figsize=(12,7))
plt.title('Top 10 Happiest Countries', fontsize=40)
sns.barplot(x=top_10.country_name, y=top_10.happiness_score, hue=top_10.regional_indicator)
plt.xlabel('Countries Name', fontsize=20)
plt.ylabel('Happiness Score', fontsize=20)
plt.xticks(rotation=30, fontsize=15, ha='right')
plt.legend(bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.title('Top 10 Highest in GDP per Capita', fontsize=40)
sns.barplot(x=top_10.country_name, y=top_10.GDP_per_capita, hue=top_10.regional_indicator)
plt.xticks(rotation=30)
plt.xlabel('Countries Name', fontsize=20)
plt.ylabel('GDP Per Capita', fontsize=20)
plt.xticks(rotation=30, fontsize=15, ha='right')
plt.legend(bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```

```python
bottom_10 = df.tail(10)
```

```python
plt.figure(figsize=(12,7))
plt.title('Bottom 10 Least Happiest Countries', fontsize=30)
sns.barplot(x=bottom_10.country_name, y=bottom_10.happiness_score, hue=bottom_10.regional_indicator)
plt.xlabel('Countries Name', fontsize=20)
plt.ylabel('Happiness Score', fontsize=20)
plt.xticks(rotation=30, fontsize=15, ha='right')
plt.legend(bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.title('Bottom 10 Least GDP per Capita', fontsize=30)
sns.barplot(x=bottom_10.country_name, y=bottom_10.GDP_per_capita, hue=bottom_10.regional_indicator)
plt.xlabel('Country Name', fontsize=20)
plt.ylabel('GDP per Capita', fontsize=20)
plt.xticks(rotation=30, fontsize=15)
plt.legend(bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.title('Top 10 Countries With Most Corruption Perception', fontsize=30)
sns.barplot(x=bottom_10.country_name, y=bottom_10.perception_of_corruption, hue=bottom_10.regional_indicator)
plt.xlabel('Country Name', fontsize=20)
plt.ylabel('Perception of Corruption', fontsize=20)
plt.xticks(rotation=30, fontsize=15, ha='right')
plt.legend(bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.title('Top 10 Countries With Highest Life Expectancy', fontsize=30)
sns.barplot(x=top_10.country_name, y=top_10.healthy_life_expectancy, hue=top_10.regional_indicator)
plt.xlabel('Country Name', fontsize=20)
plt.ylabel('Healthy Life Expactancy', fontsize=20)
plt.xticks(rotation=30, ha='right', fontsize=15)
plt.legend(bbox_to_anchor=(1.05,1))
plt.tight_layout()
plt.show()
```
